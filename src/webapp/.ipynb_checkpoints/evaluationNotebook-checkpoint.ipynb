# Let's load the datas

+ **Inspiration:** Attempt to distinguish presence with a risk factor (values 1, 2, 3, 4) of potential heart disease from absence (value 0)
+ **Features:**
    > 1. age
    > 2. sex  *(2 val possible without quantification value)*
    > 3. chest pain type (4 values)
    > 4. resting blood pressure
    > 5. serum cholestoral in mg/dl
    > 6. fasting blood sugar > 120 mg/dl *(1=true or 0=false)*
    > 7. resting electrocardiographic results *(values 0,1,2)*
    > 8. maximum heart rate achieved
    > 9. exercise induced angina
    > 10. oldpeak = ST depression induced by exercise relative to rest
    > 11. the slope of the peak exercise ST segment
    > 12. number of major vessels (0-3) colored by flourosopy
    > 13. thal: 3 = normal; 6 = fixed defect; 7 = reversable defect

+ numbers of datas are qualitative

+ not null datas
+ y output is 0 or 1 

```python
df=pd.read_csv("../../dataSetHealth/heart.csv",encoding='UTF-8')
df.columns= map(str.upper, df.columns)
df.describe()
```

## More specified informations (heatmap...)

```python
df.columns= map(str.upper, df.columns)
pdpf.ProfileReport(df)
```

# Preprocessing 

## Split datas (train and test) 

```python


y=df.TARGET
x=df.drop(['TARGET'], axis=1)

# Split the dataset to trainand test data
train_X, test_X, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=0)
```

# Random forest model

### Setup parameters

```python
parameters ={'bootstrap': True,
 'max_depth': 1,
 'max_features': 'auto',
 'max_leaf_nodes': None,
 'min_samples_leaf': 2,
 'min_samples_split': 21,
 'n_estimators': 50}
```

### Creation of the model 

```python
healthModel = RandomForestClassifier(**parameters)
```

### Training 

```python
healthModel.fit(train_X, train_y)
```

### Evaluation 

```python
RF_predictions = healthModel.predict(test_X)

score = accuracy_score(test_y ,RF_predictions)
print(score)
```

# Some stats about our forest :

## Parameters influence : 


```python
feat_labels = train_X.columns[0:]
importances = healthModel.feature_importances_ 
indices = np.argsort(importances)[::-1]
for f in range(train_X.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30,feat_labels[f], importances[indices[f]]))
```

## Best tree representation 

```python
from sklearn.externals.six import StringIO
import pydotplus
import matplotlib.image as mpimg
from sklearn import tree
%matplotlib inline 

dot_data = StringIO()
filename = "healthRandomForest1stTree.png"
df.columns=df.columns.astype(str)
featureNames = df.columns[0:13].astype(str)
targetNames = df.TARGET.astype(str).unique().tolist()
dotfilename="healthRandomForest.dot"

out=tree.export_graphviz(healthModel.estimators_[0],feature_names=featureNames, out_file=dot_data, class_names= np.unique(train_y.astype(str)))  


graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png(filename)
plt.figure(figsize=(100, 200))
img = mpimg.imread(filename)
plt.imshow(img,interpolation='nearest')
```

```python
def optimizedRandomForest(train_X, train_y,test_X,test_y,min_samples_leafMin=1, min_samples_leafMax=13, n_estimators=50, min_samples_splitMin=2, min_samples_splitMax=30, max_depthMin=1,max_depthMax=13,max_leaf_nodes=None):
    bestPrediction=0
    
    nbValmin_samples_leaf=min_samples_leafMax
    nbValmin_samples_split=min_samples_splitMax
    nbValmax_depth=max_depthMax
    nbValmax_features=3
    allVal=np.zeros((nbValmin_samples_leaf,nbValmin_samples_split,nbValmax_depth,nbValmax_features))
    
   
    for min_samples_leaf in range(min_samples_leafMin,min_samples_leafMax):
        min_samples_split=min_samples_splitMin
        for min_samples_split in range(min_samples_splitMin,min_samples_splitMax):
            max_depth=max_depthMin
            for max_depth in range(max_depthMin,max_depthMax):
            #max_feature = sqrt  == 0 
                parameters = {'bootstrap': True, #    Whether bootstrap samples are used when building trees.
              'min_samples_leaf': min_samples_leaf, # The minimum number of samples required to be at a leaf node:
              'n_estimators': n_estimators, #    The number of trees in the forest.
              'min_samples_split': min_samples_split, # The minimum number of samples required to split an internal node:
              'max_features': 'sqrt', # The number of features to consider when looking for the best split:
              'max_depth': max_depth,        # The maximum depth of the tree
              'max_leaf_nodes': max_leaf_nodes #Grow trees with max_leaf_nodes in best-first fashion
             }
                RF_model = RandomForestClassifier(**parameters) 
                RF_model.fit(train_X, train_y)
                RF_predictions = RF_model.predict(test_X)
                score = accuracy_score(test_y ,RF_predictions)
                allVal[min_samples_leaf,min_samples_split,max_depth,0]=score
                if(score>bestPrediction):
                    bestPrediction=score
                    bestParam=parameters
                    print(bestPrediction)
                    
            #max_feature = log2  == 1 
                parameters = {'bootstrap': True, #    Whether bootstrap samples are used when building trees.
              'min_samples_leaf': min_samples_leaf, # The minimum number of samples required to be at a leaf node:
              'n_estimators': n_estimators, #    The number of trees in the forest.
              'min_samples_split': min_samples_split, # The minimum number of samples required to split an internal node:
              'max_features': 'log2', # The number of features to consider when looking for the best split:
              'max_depth': max_depth,        # The maximum depth of the tree
              'max_leaf_nodes': max_leaf_nodes #Grow trees with max_leaf_nodes in best-first fashion
             }
                RF_model = RandomForestClassifier(**parameters) 
                RF_model.fit(train_X, train_y)
                RF_predictions = RF_model.predict(test_X)
                score = accuracy_score(test_y ,RF_predictions)
                allVal[min_samples_leaf,min_samples_split,max_depth,1]=score
                if(score>bestPrediction):
                    bestPrediction=score
                    bestParam=parameters
                    print(bestPrediction)

                    
             
            #max_feature = auto  == 2  
                parameters = {'bootstrap': True, #    Whether bootstrap samples are used when building trees.
              'min_samples_leaf': min_samples_leaf, # The minimum number of samples required to be at a leaf node:
              'n_estimators': n_estimators, #    The number of trees in the forest.
              'min_samples_split': min_samples_split, # The minimum number of samples required to split an internal node:
              'max_features': 'auto', # The number of features to consider when looking for the best split:
              'max_depth': max_depth,        # The maximum depth of the tree
              'max_leaf_nodes': max_leaf_nodes #Grow trees with max_leaf_nodes in best-first fashion
             }
                RF_model = RandomForestClassifier(**parameters) 
                RF_model.fit(train_X, train_y)
                RF_predictions = RF_model.predict(test_X)
                score = accuracy_score(test_y ,RF_predictions)
                allVal[min_samples_leaf,min_samples_split,max_depth,2]=score
                if(score>bestPrediction):
                    bestPrediction=score
                    bestParam=parameters
                    print(bestPrediction)

    
    return bestPrediction,bestParam, allVal
```

```python
bestPrediction,bestParam, allVal=optimizedRandomForest(train_X, train_y,test_X,test_y)
```

```python
print("the best model is : ",bestPrediction)
print("with this parameters : ",bestParam)
```

#### So, the best score with this model is 91%, it's a very good score but we will test with another model to try to get a better result

# Simple decision tree :  

```python
from sklearn.tree import DecisionTreeClassifier
healthModel = DecisionTreeClassifier(criterion="entropy", max_depth = 7) 
healthModel.fit(train_X, train_y)

RF_predictions = healthModel.predict(test_X)

score = accuracy_score(test_y ,RF_predictions)
print(score)
```

```python
from sklearn.externals.six import StringIO
import pydotplus
import matplotlib.image as mpimg
from sklearn import tree
%matplotlib inline 

dot_data = StringIO()
filename = "healthDecisionTree.png"
df.columns=df.columns.astype(str)
featureNames = df.columns[0:13].astype(str)
targetNames = df.TARGET.astype(str).unique().tolist()
dotfilename="healthDecisionTree.dot"

out=tree.export_graphviz(healthModel,feature_names=featureNames, out_file=dot_data, class_names= np.unique(train_y.astype(str)))  


graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png(filename)
plt.figure(figsize=(100, 200))
img = mpimg.imread(filename)
plt.imshow(img,interpolation='nearest')
```
